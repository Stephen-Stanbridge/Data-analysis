```python
temperature_df = pd.read_csv("land_temperatures/GlobalTemperatures.csv")
pacific_wind_df = pd.read_csv("hurricanes/atlantic.csv")
```

<font size=4>Create function for changing date format of winds dataframe:</font><br><br>

```python
def change_hurricane_date_format(date):
    year = date[:4]
    month = date[4:6]
    day = date[6:9]
    return f"{year}-{month}-{day}"
```

<font size="4">Mapping function to every date in 'Date' column.</font><br><br>

```python
pacific_wind_df["Date"] = pacific_wind_df["Date"].astype(str).apply(change_hurricane_date_format)
```

<font size="4">Listing all winds above 117km/h which means that they were hurricanes.</font><br><br>

```python
pacific_wind_df[(pacific_wind_df["Maximum Wind"] > 117)]
```

<font size="3"> *-999 values are the way of representing cells that were not measured.</font><br>

<font size="4">Change name of 'dt' column.</font>

```python
temperature_df.rename(columns={"dt": "Date"}, inplace=True)
```

<font size="4">Select only needed columns for data comparision.</font>

```python
pacific_wind_df = pacific_wind_df[['Date', 'Maximum Wind']]
```

<font size="4">Select max wind measure for every month since 1851-06.</font>

```python
values = []
date = "1851-06"
columns = {"Date": [], "Max_wind": []}
for row in pacific_wind_df.itertuples():
    if date != row[1][:7]:
        columns["Date"].append(date)
        columns["Max_wind"].append(max(values)) if len(values)!=0 else columns["Max_wind"].append(np.NaN)
        values = []
        date = row[1][:7]
    values.append(row[2])
max_wind_every_month_df = pd.DataFrame(columns)
```

```python
max_wind_every_month_df.head()
```

<font size="4">Function for changing date format so it shows only year and month. Day is useless since average temperature measurement has been taken every month.</font>

```python
def change_temperature_date_format(date):
    return date[:7]
```

<font size="4">Select only needed columns for data comparision.</font>

```python
temperature_df = temperature_df[["Date", "LandAverageTemperature"]]
```

```python
temperature_df.head()
```

<font size="4">Apply changing date function so both dataframes have same date format.</font>

```python
temperature_df["Date"] = temperature_df["Date"].apply(change_temperature_date_format)
```

```python
temperature_df.rename(columns={"LandAverageTemperature": "Average_temperature"}, inplace=True)
```

<font size="4"> Take temperature measurements from 1851-06 so it has same timeline as winds dataframe.</font>

```python
temperature_df = temperature_df[temperature_df["Date"] >= "1851-06"]
```

<font size="4"> Set same indexes for both dataframes so they can be merged.</font>

```python
temperature_df.set_index("Date", inplace=True)
max_wind_every_month_df.set_index("Date", inplace=True)
```

```python
temperature_hurricane_df = temperature_df.merge(max_wind_every_month_df, left_index=True, right_index=True)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

<font size="4">Function for creating labels every 10 years so it looks tidy on the plot.</font>

```python
def create_labels_for_years():
    labels = []
    last_date = "185"
    for date in temperature_hurricane_df.index:
        if date[:3] != last_date:
            last_date = date[:3]
            labels.append(last_date + "0")
            continue
        labels.append("")
    return labels
```

<font size="4"> Function for getting X set of winds that are above certain velocity.</font>

```python
def get_xs_of_winds(speed):
    xs = []
    for row in temperature_hurricane_df.reset_index().iterrows():
        if row[1]["Max_wind"] > speed:
            xs.append(row[0])
    return xs
```

<font size="4"> Function for calculationg best fit slope and intercept that will be used in calculating regression line.</font>

```python
from statistics import mean
def calculate_best_fit_slope_and_intercept(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    m = (mean(xs) * mean(ys) - mean(xs*ys)) / (mean(xs)**2 - mean(xs**2))
    b = mean(ys) - m*mean(xs)
    return {"m": m, "b": b}
```

```python
labels = create_labels_for_years()

# Create appropriate x axis length
X = np.arange(len(labels))

# Create subplot and figure
figure, axis = plt.subplots()
figure.set_size_inches(23,11)

# Temperature plot
# Place labels in appropriate locations
plt.xticks(X, labels, rotation=60)

# Plot temperatures, take mean of every 10 measurements
axis.plot(X, temperature_hurricane_df['Average_temperature'].rolling(10).mean(), label="Temperature", c="orange")

# Set label for Y axis
axis.set_ylabel("Celcius degrees", fontsize=14)

# Calculate best fit slope and intercept from all temperatures and plot regression line
regression_line_for_temperatures = calculate_best_fit_slope_and_intercept(X, temperature_hurricane_df['Average_temperature'].values)
axis.plot(X, regression_line_for_temperatures['m'] * X + regression_line_for_temperatures['b'], label="Regression line of temperatures", c="green")

# Wind plot
# Scatter winds that are above 117km/h which means that they were classified as hurricanes
xs = get_xs_of_winds(117)

# Place all hurricanes on same y value so it looks tidy but informational about when and how many hurricanes were in certain period of time
axis.scatter(xs, [2.5 for x in xs], c='red', s=30, label="Hurricanes")

# Storm plot
# Plot winds
xs = get_xs_of_winds(0)

# Create twin axis so both plots share it
wind_velocity_axis = axis.twinx()
wind_velocity_axis.plot(xs, temperature_hurricane_df['Max_wind'][temperature_hurricane_df['Max_wind'] > 0].rolling(10).mean(), label='Wind velocity', c="lightblue")


# Wind regression line plot
regression_line_for_wind_velocity = calculate_best_fit_slope_and_intercept(X, temperature_hurricane_df['Max_wind'].values)
wind_velocity_axis.plot(X, regression_line_for_wind_velocity['m'] * X + regression_line_for_wind_velocity['b'], label="Regression line of all wind velocities", c="lightgreen")

# Set limit for y axis so graph looks tidy but informative
wind_velocity_axis.set_ylim(0,250)
wind_velocity_axis.set_ylabel("Km/h", fontsize=14)

figure.legend()
figure.tight_layout()
plt.show()
```
